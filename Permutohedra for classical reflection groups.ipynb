Creating a type $A$ permutohedron by taking the convex hull of the orbit of an identity element.   We can label the alcoves, but the resulting plot looks pretty cluttered.

```python
V = RootSystem(["A",2]).ambient_space()
A3 = V.weyl_group()
e = V.basis()
Lambda = V.fundamental_weights()
id = e[0] + 2*e[1] + 3*e[2]
S3 = [w.action(id) for w in A3]
```

```python
plot_opt = V.plot_parse_options()
verts = [1.3*plot_opt.projection(s) for s in S3]

plt = V.plot(alcoves=True, alcove_labels=True)
Perm_A3 = Polyhedron(vertices = verts)

plt = plt + Perm_A3.plot(color="lightblue")
show(plt, axes=false)
```

The is a type $D$ permutohedron created by taking the convex hull of a strictly ordered point with $n-1$ support. Why do we require this smaller support?

```python
L = RootSystem(["D",3]).ambient_space()
rho = L.rho()
print("The orbit of: " + str(rho))
plot_options = L.plot_parse_options()
W = L.weyl_group()
g = W.cayley_graph(side="right")
positions = [plot_options.projection(w.action(rho)) for w in W]


plt = L.plot(alcoves=True)
perm_D3 = Polyhedron(vertices = positions) 
plt += perm_D3.plot()
show(plt, axes=false)
```
